# Checking linear separability through convex hull

```python
import numpy as np
from scipy.spatial import ConvexHull, convex_hull_plot_2d
import matplotlib.pyplot as plt
```

```python
X = np.array([
    [0, 1], 
    [0, 0], 
    [1, 1],
    [-1, 0], 
    [1, -1], 
    [-1, -1]
])
y = np.array([1, 1, 1, -1, -1, -1])
```

```python
hull_pos = ConvexHull(X[:3])
hull_neg = ConvexHull(X[3:])
```

```python
plt.plot(X[:3][:,0], X[:3][:,1], 'o')
plt.plot(X[3:][:,0], X[3:][:,1], 'x')
for simplex in hull_pos.simplices:
    plt.plot(X[:3][simplex, 0], X[:3][simplex, 1], 'k-')
for simplex in hull_neg.simplices:
    plt.plot(X[3:][simplex, 0], X[3:][simplex, 1], 'k-')
```

Hulls not overlapping --> the problem is linearly separable
